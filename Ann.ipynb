```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df= pd.read_csv('cancer.csv')
```

```python
df
```

```python
# counting values of variables in 'diagnosis'
df['diagnosis'].value_counts()
```

```python
plt.figure(figsize=[5,5])
#sb.countplot(df['diagnosis'].value_counts())
sb.countplot(x ='diagnosis', data = df)
plt.show()
```

```python
df.isnull().sum()
```

```python
#droping feature
df.drop(['Unnamed: 32','id'],axis=1,inplace=True)
```

```python
# independent variables
x = df.drop('diagnosis',axis=1)
#dependent variables
y = df.diagnosis
```

```python
x
```

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder
#creating the object
lb = LabelEncoder()
y = lb.fit_transform(y)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.3,random_state=40)
```

```python
#importing StandardScaler
from sklearn.preprocessing import StandardScaler
#creating object
sc = StandardScaler()
xtrain = sc.fit_transform(xtrain)
xtest = sc.transform(xtest)
```

```python
xtrain
```

```python
xtest
```

```python
#importing keras
import keras
#importing sequential module
from keras.models import Sequential
# import dense module for hidden layers
from keras.layers import Dense
from keras.utils import plot_model
#importing activation functions
#from keras.layers import ReLU
#from keras.layers import Dropout
```

```python
#creating model
classifier = Sequential()
```

```python
#first hidden layer
classifier.add(Dense(units=12,activation='relu',input_dim=30))
#second hidden layer
classifier.add(Dense(units=12,activation='relu'))
# last layer or output layer
classifier.add(Dense(units=1,activation='sigmoid'))
```

```python
plot_model(classifier,show_shapes=True)
```

```python
#taking summary of layers
classifier.summary()
```

```python
#compiling the ANN
classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
#fitting the ANN to the training set
model = classifier.fit(xtrain,ytrain,batch_size=100,epochs=200)
```

```python
#now testing for Test data
y_pred = classifier.predict(xtest)
```

```python
y_pred
```

```python
#converting values
y_pred = (y_pred>0.5)
print(y_pred)
```

```python
ytest
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
cm = confusion_matrix(ytest,y_pred)
score = accuracy_score(ytest,y_pred)
print(cm)
print('score is:',score)
```

```python
# creating heatmap of comfussion matrix
plt.figure(figsize=[14,7])
sb.heatmap(cm,annot=True)
plt.show()
```

```python
# list all data in history
print(model.history.keys())
# summarize history for accuracy
plt.plot(model.history['accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
# summarize history for loss
plt.plot(model.history['loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```
